```python
import numpy as np
import h5py
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from sympy import isprime

# ================================================================
# 1. Load Prime-Encoded Frequency Differences from All Observing Runs & Detectors
# ================================================================
# List of datasets (O1, O2, O3 - H1, L1, V1)
datasets = {
    "O2_L1": r"C:\Users\walte\gw_data\L-L1_GWOSC_O2_4KHZ_R1-1181261824-4096.hdf5",
    "O3_V1": r"C:\Users\walte\gw_data\V-V1_GWOSC_O3a_4KHZ_R1-1238167552-4096.hdf5"
}

dataframes = {}
for label, file in datasets.items():
    try:
        with h5py.File(file, 'r') as f:
            if 'strain' in f.keys():
                strain_data = f['strain']['Strain'][:]
                dataframes[label] = strain_data
                print(f"âœ… Loaded {label}: {file}")
            else:
                print(f"âŒ Error: 'strain' dataset not found in {label}")
    except Exception as e:
        print(f"âŒ Error loading {label}: {e}")

# ================================================================
# 2. Compute FFT & Identify Peaks in Each Dataset
# ================================================================
sampling_rate = 4096  # LIGO standard

plt.figure(figsize=(10, 6))
for label, strain_data in dataframes.items():
    N = len(strain_data)
    fft_values = np.abs(np.fft.fft(strain_data))
    frequencies = np.fft.fftfreq(N, d=1/sampling_rate)
    
    mask = (frequencies > 0)  # Keep only positive frequencies
    plt.plot(frequencies[mask], fft_values[mask], label=label)

plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.title("FFT Spectrum Across LIGO Runs & Detectors")
plt.legend()
plt.grid(True)
plt.show()

# ================================================================
# 3. Extract Prime-Encoded Frequency Differences
# ================================================================
def extract_prime_intervals(frequencies, fft_values):
    peaks, _ = find_peaks(fft_values, height=np.mean(fft_values) + 2 * np.std(fft_values))
    peak_freqs = frequencies[peaks]
    freq_diffs = np.diff(peak_freqs)
    prime_diffs = [diff for diff in freq_diffs if isprime(int(round(diff)))]
    return prime_diffs, freq_diffs

for label, strain_data in dataframes.items():
    N = len(strain_data)
    fft_values = np.abs(np.fft.fft(strain_data))
    frequencies = np.fft.fftfreq(N, d=1/sampling_rate)

    prime_diffs, freq_diffs = extract_prime_intervals(frequencies, fft_values)
    print(f"Prime-Encoded Frequency Differences for {label}: {prime_diffs}")

# ================================================================
# 4. Monte Carlo Validation for Prime Structure
# ================================================================
def monte_carlo_prime_test(freq_diffs, num_iterations=100000):
    num_diffs = len(freq_diffs)
    random_matches = 0
    for _ in range(num_iterations):
        random_freqs = sorted(np.random.randint(10, 2000, num_diffs))
        random_diffs = np.diff(random_freqs)
        prime_count = sum(1 for diff in random_diffs if isprime(int(round(diff))))
        if prime_count >= len(freq_diffs):
            random_matches += 1
    return random_matches / num_iterations

for label, strain_data in dataframes.items():
    N = len(strain_data)
    fft_values = np.abs(np.fft.fft(strain_data))
    frequencies = np.fft.fftfreq(N, d=1/sampling_rate)

    _, freq_diffs = extract_prime_intervals(frequencies, fft_values)
    p_value = monte_carlo_prime_test(freq_diffs)
    print(f"Monte Carlo p-value for {label}: {p_value:.6f}")
    if p_value < 1e-6:
        print(f"ðŸš€ Statistically significant prime-number encoding detected in {label}!")
    else:
        print(f"No strong evidence of structured prime encoding in {label}.")
```
